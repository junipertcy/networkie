# Visualization

Main goals:
1. Learn how to visualize a network
2. Learn the constraints in network visualization
3. Learn that the visualization can bias our understanding to the system

我們先讀進網路的數據：

```python
import networkx as nx
from networkie.gen import Custom
lff = Custom.LoadFromFile()
g = lff.from_edgelist('dataset/karate.edgelist')
```

我們可以試看看不同的 graph layout 效果，不同的效果會把節點擺在平面上的不同位置。注意不同的效果只是視覺上的差異，不應該對視覺畫的結果作過多的解讀（例如兩節點畫出來比較近，不一定代表他們在現實中就比較相近）。

```python
nx.draw_circular(g, with_labels=True, node_color='#7FDBFF', edge_color='grey', alpha=0.8)
```

```python
nx.draw_spring(g, with_labels=True, node_color='#7FDBFF', edge_color='grey', alpha=0.8)
```

如果過度的對網路視覺化的結果推論，可能會有錯誤的理解。如這個例子：https://twitter.com/karlrohe/status/977267179258294272
